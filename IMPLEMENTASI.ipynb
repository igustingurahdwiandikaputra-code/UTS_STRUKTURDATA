```python
class AntrianRS:
    def __init__(self, kapasitas):
        self.queue = []
        self.kapasitas = kapasitas

    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, nama_pasien):
        """Menambah pasien ke antrian"""
        if len(self.queue) < self.kapasitas:
            self.queue.append(nama_pasien)
            print(f"[Input] Pasien {nama_pasien} berhasil masuk antrian.")
        else:
            print("Antrian penuh!")

    def dequeue(self):
        """Melayani (menghapus) pasien terdepan"""
        if self.is_empty():
            print("Antrian kosong, tidak ada pasien untuk dilayani.")
        else:
            pasien = self.queue.pop(0)
            print(f"[Proses] Pasien {pasien} selesai dilayani dan keluar dari antrian.")

    def peek(self):
        """Melihat pasien urutan pertama"""
        if not self.is_empty():
            print(f"[Info] Pasien urutan terdepan: {self.queue[0]}")
        else:
            print("Antrian kosong.")

    def display(self):
        """Menampilkan seluruh antrian"""
        if self.is_empty():
            print("[Output] Daftar Antrian: Kosong")
        else:
            print(f"[Output] Daftar Antrian saat ini: {' -> '.join(self.queue)}")

```

UJI COBA SISTEM

```python
rs_sehat = AntrianRS(5)

rs_sehat.enqueue("Riska")
rs_sehat.enqueue("Radit")
rs_sehat.enqueue("Dwi")
rs_sehat.display()

rs_sehat.peek()
rs_sehat.dequeue()
rs_sehat.display()
```
